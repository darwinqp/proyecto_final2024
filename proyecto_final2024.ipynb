```python
# Importamos los paquetes
import os 
import pyreadstat
import pickle
import pandas as pd
import numpy as np
import urllib.request
import altair as alt
import wbgapi as wb
```

```python
# Importamos la base de datos, en este caso de los indicadores de Desarrollo GlobaL, a nivel país. 

os.getcwd()
base = pd.read_csv("./data/51f7ab6a-24ed-4a06-9f5d-04873f1673fd_Data-1.csv")
```

```python
# Borramos todo lo que no nos interesa, las celdas finales y los YEARS
base.columns = base.columns.str.replace(r'\[.*', '', regex=True)
base = base.drop(list(range(1596, 1601)))
base
```

```python
# Seleccionamos el pais en el cual nos queremos enfocar: en este caso Peru
pais_seleccionado = 'Peru'
base_pais = base[base['Country Name'] == pais_seleccionado]
base_pais
```

```python
# Borramos lo que no nos interesa y lo transponemos
base_pais = base_pais.drop(['Country Name', 'Country Code', 'Series Code'], axis=1)
base_pais = base_pais.transpose()

# Crear un DataFrame con los datos proporcionados
base_pais = pd.DataFrame(base_pais)

# Tomar la primera fila como encabezados
base_pais.columns = base_pais.iloc[0]

# Eliminar la primera fila que ahora es redundante
base_pais = base_pais[1:]

base_pais

```

```python
# Extraemos los años del index y con ellos creamos otra variable llamada año

base_pais.reset_index(inplace=True)
base_pais.rename(columns={'index': 'año'}, inplace=True)

# Remover las filas que contienen '..' en los valores, ya que Altair espera datos numéricos
df_numeric = base_pais.replace('..', pd.NA).dropna()

base_pais
```

```python
# Crear el gráfico primer grafico de la Evolicion del Crecimeinto del PIB

chart = alt.Chart(base_pais).mark_line().encode(
    x='año:T',
    y=alt.Y('GDP growth (annual %):Q', title='Crecimiento del PIB (%)'),  # Personalizar el título del eje Y
    color=alt.value('blue'),  # Cambiar el color de la línea
).properties(
    width=800,  # Ajustar el ancho del gráfico
    height=400,  # Ajustar la altura del gráfico
    title='Evolución del Crecimiento del PIB',  # Añadir un título al gráfico
).configure_axis(
    labelFontSize=12,  # Tamaño de la fuente de las etiquetas del eje
    titleFontSize=14,  # Tamaño de la fuente del título del eje
).configure_title(
    fontSize=16,  # Tamaño de la fuente del título del gráfico
)

# Mostrar el gráfico
chart
```

```python
# Tanteamos algunos graficos

# Filtrar las columnas necesarias
df_subset =  base_pais[['año', 'Access to electricity (% of population)', 'GDP growth (annual %)']].copy()

# Convertir las columnas a formato numérico, reemplazando '..' con NaN
df_subset['Access to electricity (% of population)'] = pd.to_numeric(df_subset['Access to electricity (% of population)'], errors='coerce')
df_subset['GDP growth (annual %)'] = pd.to_numeric(df_subset['GDP growth (annual %)'], errors='coerce')

# Crear el gráfico de línea
chart = alt.Chart(df_subset).mark_line().encode(
    x='año:T',
    y=alt.Y(alt.repeat("column"), type='quantitative'),
    color='Indicador:N'
).properties(
    width=600,
    height=400
).repeat(
    column=['Access to electricity (% of population)', 'GDP growth (annual %)'],
    title=['Access to electricity (% of population)', 'GDP growth (annual %)']
).configure_title(
    anchor='middle'
)

chart
```

```python
# decidimos comparar la evolicion del acceso a la elecctricidad y la fuerza laboral femenina

base_pais['año'] = pd.to_numeric(base_pais['año'], errors='coerce')

# Filtrar las filas para considerar solo los años entre 1990 y 2022
df_subset = base_pais[(base_pais['año'] >= 1990) & (base_pais['año'] <= 2022)][['año', 'Access to electricity (% of population)', 'Labor force, female (% of total labor force)']].copy()

# Convertir las columnas a formato numérico, reemplazando '..' con NaN
df_subset['Access to electricity (% of population)'] = pd.to_numeric(df_subset['Access to electricity (% of population)'], errors='coerce')
pd.to_numeric(df_subset['Labor force, female (% of total labor force)'], errors='coerce')

# Derretir el DataFrame para facilitar el gráfico
df_melted = df_subset.melt(id_vars=['año'], var_name='Indicador', value_name='Valor')

# Crear el gráfico de línea
chart = alt.Chart(df_melted).mark_line().encode(
    x='año:T',
    y='Valor:Q',
    color='Indicador:N'
).properties(
    width=600,
    height=400,
    title='Evolución de Access to electricity y Fuerza laboral femenina entre 1990 al 2022'
)

chart
```

```python
# Correcion de las fechas anteriores
# Convertir la columna 'año' a tipo fecha
df_subset['año'] = pd.to_datetime(df_subset['año'], format='%Y')

# Derretir el DataFrame para facilitar el gráfico
df_melted = df_subset.melt(id_vars=['año'], var_name='Indicador', value_name='Valor')

# Crear el gráfico de línea
chart = alt.Chart(df_melted).mark_line().encode(
    x='año:T',  # Usar 'T' para indicar que se trata de datos temporales
    y='Valor:Q',
    color='Indicador:N'
).properties(
    width=600,
    height=400,
    title='Evolución de Access to electricity y de la Fuerza laboral femenina entre 1990 y 2022'
)

# Mostrar el gráfico
chart
```
